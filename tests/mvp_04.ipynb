```python
# Delte columns where all values are NaN. There some columns with no data because the file was made
# acording to the Darwin Core standard: https://dwc.tdwg.org/terms/
df.dropna(how='all', axis=1, inplace=True)
```

```python
df = df[~df['species'].isna()]
```

```python
df['locality'].isna().sum()
```

```python
df[(df['decimalLatitude'] == 0) & (df['decimalLongitude'] == 0)].shape
```

```python
df = df[~((df['decimalLatitude'] == 0) & (df['decimalLongitude'] == 0))]
```

```python
df.shape
```

```python
df['locality'].isna().sum()
```

```python
df.sort_values(by=['decimalLongitude','decimalLatitude'], inplace=True)
```

```python
df['locality2'] = df['locality'].fillna(method='ffill')
```

```python
df[['waterBody','countryCode','kingdom','phylum','class','order','family','genus']].isna().sum()
```

```python
df['waterBody2'] = df['waterBody'].fillna(method='ffill')
```

```python
df[['waterBody2','countryCode','kingdom','phylum','class','order','family','genus']].isna().sum()
```

```python
df['class2'] = df['class'].fillna(value='unknown')
df['order2'] = df['order'].fillna(value='unknown')
```

```python
num_columns = ['month','day','depth']
cat_columns = ['waterBody2','countryCode','kingdom','phylum','class2','order2','family','genus']
df[cat_columns + num_columns].isna().sum()
```

```python
max_species = df['species'].str.strip().value_counts().index[0]
df['target'] = (df['species'] == max_species).astype(int)
```
